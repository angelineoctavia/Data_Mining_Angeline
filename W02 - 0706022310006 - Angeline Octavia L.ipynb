```python
import pandas as pd

# Download the CSV file from the URL
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url, sep=',')

df.head()
```

```python
# Untuk menunjukkan ringkasan struktur data,
# seperti column, non-null count dan data type.

df.info()
```

```python
# Menampilkan beberapa kolom numerik seperti price_in_rp, lat, long dan sebagainya.
# Dari data di bawah ini, dapat dilihat bahwa terdapat beberapa data
# yang kurang masuk akal (outlier), seperti jumlah bedroom hingga 3519 dan 99,
# dan jumlah bathrooms sebesar 3524. 

df.describe()
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]

# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)

null_df

# Untuk menghitung jumlah null beserta presentasenya.
```

```python
# Menggunakan median agar untuk mengisi missing values karena median tidak terpengaruh 
# oleh outlier sehingga hasil imputasi lebih representatif. Selain itu, median menjaga 
# distribusi data numerik yang umumnya skewed seperti harga rumah agar tetap stabil.

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

df.drop(columns=['url', 'ads_id', 'address'], inplace=True)

# Memastikan tidak ada yang null yang tersisa
df.isnull().sum()
```

```python
# Boxplot menunjukkan bahwa sebagian besar rumah memiliki jumlah 
# kamar tidur, kamar mandi, dan garasi dalam rentang normal, sementara harga, 
# luas tanah, dan bangunan masih ada beberapa nilai yang lebih tinggi dari mayoritas 
# meskipun nilai ekstrem sudah dibatasi. Visualisasi ini membantu memastikan analisis 
# selanjutnya lebih akurat dan tidak terlalu dipengaruhi nilai ekstrem.

import seaborn as sns
import matplotlib.pyplot as plt

df = df[df['land_size_m2'] <= 300]  
df = df[df['building_size_m2'] <= 8000]
df = df[df['price_in_rp'] <= 5_000_000_000]

# Boxplot: Price
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price_in_rp'])
plt.title('Boxplot Harga Rumah (price_in_rp)')
plt.xlabel('Harga Rumah (Rp)')
plt.show()

# Boxplot: Bedrooms
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bedrooms'])
plt.title('Boxplot Jumlah Kamar Tidur (bedrooms)')
plt.xlabel('Jumlah Kamar Tidur')
plt.show()

# Boxplot: Bathrooms
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bathrooms'])
plt.title('Boxplot Jumlah Kamar Mandi (bathrooms)')
plt.xlabel('Jumlah Kamar Mandi')
plt.show()

# Boxplot: Garages
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['garages'])
plt.title('Boxplot Jumlah Garages')
plt.xlabel('Jumlah Garages')
plt.show()

# Boxplot: Luas Tanah (land_size_m2)
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['land_size_m2'])
plt.title('Boxplot Luas Tanah (m2)')
plt.xlabel('Luas Tanah (m2)')
plt.show()

# Boxplot: Luas Bangunan (building_size_m2)
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['building_size_m2'])
plt.title('Boxplot Luas Bangunan (m2)')
plt.xlabel('Luas Bangunan (m2)')
plt.show()
```

```python
# Ini memudahkan analisis dan perbandingan karakteristik rumah berdasarkan kelompok harga.

bins = [0, 500_000_000, 1_000_000_000, 2_000_000_000, df['price_in_rp'].max()]
labels = ['Low', 'Medium', 'High', 'Luxury']

df['Price_Category'] = pd.cut(df['price_in_rp'], bins=bins, labels=labels)

df[['price_in_rp', 'Price_Category']].head()
```

## Part B â€“ Exploratory Data Analysis (EDA)

```python
# Kategori Harga Rumah (Low, Medium, High, Luxury)

# Berdasarkan visualisasi jumlah rumah menurut kategori harga, 
# terlihat bahwa kategori Luxury memiliki jumlah properti terbanyak, 
# diikuti oleh kategori High dan Medium, sementara kategori Low memiliki 
# jumlah rumah paling sedikit. Hal ini menunjukkan bahwa sebagian besar 
# rumah di dataset termasuk dalam kategori harga tinggi, sehingga analisis 
# atau strategi terkait pasar properti premium akan lebih relevan.

plt.figure(figsize=(6,4))
sns.countplot(x='Price_Category', data=df, order=['Low','Medium','High','Luxury'])
plt.title("Jumlah Rumah Berdasarkan Kategori Harga")
plt.xlabel("Kategori Harga")
plt.ylabel("Jumlah Properti")
plt.show()
```

```python
# Scatterplot: Luas Bangunan vs Harga Rumah
# Scatterplot menunjukkan hubungan positif antara luas bangunan dan harga 
# rumah: semakin besar rumah, semakin tinggi harganya.

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='building_size_m2', y='price_in_rp', alpha=0.5)
plt.ylim(0, 10_000_000_000)  # batasi supaya pola lebih jelas
plt.title("Hubungan Luas Bangunan dan Harga Rumah")
plt.xlabel("Luas Bangunan (m2)")
plt.ylabel("Harga Rumah (Rp)")
plt.show()
```

```python
# Heatmap korelasi harga rumah dengan faktor utama

# Heatmap menunjukkan korelasi antara harga rumah dan faktor numerik 
# lainnya. Terlihat bahwa harga rumah memiliki korelasi positif 
# paling kuat dengan luas bangunan dan luas tanah, sementara faktor 
# lain cenderung lebih lemah. Ini menandakan bahwa ukuran rumah dan 
# tanah adalah prediktor utama harga rumah dalam dataset ini.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# Distribusi Luas Tanah

# key findings
# Distribusi luas tanah menunjukkan sebagian besar properti memiliki 
# ukuran tanah kecil hingga sedang, dengan jumlah properti menurun 
# seiring bertambahnya luas tanah. Ada beberapa properti dengan tanah 
# sangat luas, namun jarang, sehingga terlihat sebagai outlier.

plt.figure(figsize=(8,5))
sns.histplot(df['land_size_m2'], bins=50, kde=True)
plt.title("Distribusi Luas Tanah Properti")
plt.xlabel("Luas Tanah (m2)")
plt.ylabel("Jumlah Properti")
plt.show()
```

```python
# Distribusi harga rumah berdasarkan kota

# Beberapa kota memiliki median harga lebih tinggi, menunjukkan 
# properti di sana umumnya lebih mahal, sementara kota lain lebih 
# terjangkau. Terdapat juga outlier dengan harga sangat tinggi 
# di beberapa kota, menandakan adanya rumah mewah yang jarang.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(x='city', y='price_in_rp', data=df, palette='Set2')
plt.title('Distribusi Harga Rumah berdasarkan Kota')
plt.xlabel('Kota')
plt.ylabel('Harga Rumah (Rp)')
plt.xticks(rotation=45)
plt.show()
```
